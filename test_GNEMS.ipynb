```python
from GNEMS.GNEMS import GNEMS_segment
from PIL import Image
from PIL.PngImagePlugin import PngInfo
import numpy as np
import matplotlib.pyplot as plt
from skimage.segmentation import mark_boundaries
from skimage.color import label2rgb
from glob import glob
from tqdm import tqdm
import time
```

```python
with open("datasets/clouds/test_ids.txt", "r") as f:
    test_ids = [line.strip() for line in f.readlines()]

for id in tqdm(test_ids):
    path = f"datasets/clouds/images/{id}.png"
    im = np.array(Image.open(path).convert("L"))
    start_time = time.time()
    seg = GNEMS_segment(im, d=15, iterations=50, prediction_stride=8)
    segmentation_time = time.time() - start_time
    # fig, ax = plt.subplots(1, 2, figsize=(6, 3))
    # ax[0].imshow(im)
    # ax[1].imshow(seg)
    # plt.show()
    metadata = PngInfo()
    metadata.add_text("segmentation_time", str(segmentation_time))
    Image.fromarray(seg).convert("L").save(f"results/GNEMS/clouds/{id}.png", pnginfo=metadata)

with open("datasets/noise/test_ids.txt", "r") as f:
    test_ids = [line.strip() for line in f.readlines()]

for id in tqdm(test_ids):
    path = f"datasets/noise/{id}"
    im = np.array(Image.open(path).convert("L"))
    start_time = time.time()
    seg = GNEMS_segment(im, iterations=50, prediction_stride=8)
    segmentation_time = time.time() - start_time
    # fig, ax = plt.subplots(1, 2, figsize=(6, 3))
    # ax[0].imshow(im)
    # ax[1].imshow(seg)
    # plt.show()
    metadata = PngInfo()
    metadata.add_text("segmentation_time", str(segmentation_time))
    result = Image.fromarray(seg).convert("L")
    result.save(f"results/GNEMS/noise/{id}", pnginfo=metadata)

with open("datasets/texture/test_ids.txt", "r") as f:
    test_ids = [line.strip() for line in f.readlines()]

for id in tqdm(test_ids):
    path = f"datasets/texture/images/{id}.png"
    im = np.array(Image.open(path).convert("L"))
    start_time = time.time()
    seg = GNEMS_segment(im, iterations=50, prediction_stride=8)
    segmentation_time = time.time() - start_time
    # fig, ax = plt.subplots(1, 2, figsize=(6, 3))
    # ax[0].imshow(im)
    # ax[1].imshow(seg)
    # plt.show()
    metadata = PngInfo()
    metadata.add_text("segmentation_time", str(segmentation_time))
    Image.fromarray(seg).convert("L").save(f"results/GNEMS/texture/{id}.png", pnginfo=metadata)
```
